```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X = gdp_df[['GDP']].values 

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

sse = []
k_range = range(1, 8)  
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Број На Кластери')
plt.ylabel('SSE')
plt.title('Метод За Утврдување На Оптимален Број Кластери При БДП По Глава На Жител')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
gdp_df['Cluster'] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=gdp_df, x='Country', y='GDP', hue='Cluster', palette='Set1')
plt.title('Кластери На БДП По Глава На Жител 2023')
plt.xlabel('Земја')
plt.ylabel('БДП По Глава На Жител')
plt.xticks(rotation=90)
plt.show()
```
